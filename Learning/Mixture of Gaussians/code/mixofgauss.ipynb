```python
%matplotlib inline
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
latents = ss.multinomial.pmf(x = [[0,1],[1,0]], n=[1,1], p =[0.7,0.3] )
observed= np.linspace(-4,10,1000)
likhood1= ss.norm.pdf(observed,loc= 0, scale =1)
likhood2= ss.norm.pdf(observed, loc =6, scale =2)
p_x     = 0.7*likhood1 + 0.3*likhood2
num1= (0.7*likhood1) 
den = (0.3*likhood1 + 0.7*likhood2)
num2 = (0.3*likhood2)
posterior1= num1/den
posterior2=num2/den
fig, ax = plt.subplots()
ax.plot(observed, p_x,'r', label='marginal')
ax.plot(observed, likhood1, 'b', label = 'likelihood1')
ax.plot(observed, likhood2, 'k', label = 'likelihood2')
ax.plot(observed, posterior1, 'g', label = 'Posterior z=1')
ax.plot(observed, posterior2, label = 'Posterior z=2')
plt.legend()
plt.ylim([0,0.5])
ax.grid(True)
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
plt.savefig('example1.pdf')
plt.show()


```
